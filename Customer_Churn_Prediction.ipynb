# Handling Imbalanced dataset with SMOTE

```python
# using oversampling
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X = data.drop('Exited',axis=1)
y = data['Exited']
```

```python
X_result, y_result = SMOTE().fit_resample(X,y)
```

# Significant factors contributing to churn

```python
import pandas as pd

importances = model.feature_importances_
feature_names = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard', 
                 'IsActiveMember', 'EstimatedSalary', 'Geography_Germany', 'Geography_Spain', 'Gender']


importance_df = pd.DataFrame(list(zip(feature_names, importances)), columns=['Feature', 'Importance'])
importance_df = importance_df.sort_values(by='Importance', ascending=False)
print(importance_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='Green')
plt.xlabel('Importance')
plt.title('Feature Importance - Random Forest Model')
plt.gca().invert_yaxis()
plt.show()
```

### From the above visualization we get to know that the most significant features for churn are: Age, EstimatedSalary, Balance and CreditScore
